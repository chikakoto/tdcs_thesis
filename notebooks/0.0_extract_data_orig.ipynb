# B (magnetic field) = Gyromagnetic ratio (constant value) * W (frequency)
## calculating W will give B

# W = delta phase / delta time 
## 1. Raw data 
### 12 data point (time) of delta phase 
### [mag1, phase1]

## 2. Radient 
### Pi / 4096
### phase1_rad
### [phase2_m_1_rad]

## 3. Unwrap 
### 11 data point (time - previous time) 
### Average them 
### [phase2_m_1_unwr_rad_np, fmap2_m_1_nT]
### [fmap_Mean]

## 4. Get actual and theory value
### Phase unwrap ON - Phase unwrap OFF 
### [diff_fmap_Mean, Bval]

## 12 data point
### mag, raw data & radient

## B value
### fmap mean, fmap diff, B value

```python
new_data = []
for m in range(len(base_runs)):
    bdir = cwd_path+"base_run_"+base_runs[m]+"/"
    tdir = cwd_path+bp[m][0]+"minus"+bp[m][1]+"/"
    
    theory_img = nib.load(tdir+"Bvals.nii")
    theory_data = theory_img.get_data()
    sp = theory_data.shape
    
    fmap_img = nib.load(bdir+"fmap_Mean.nii.gz")
    fmap_data = fmap_img.get_data()
    
    fmap_diff_img = nib.load(tdir+"diff_fmap_Mean.nii.gz")
    fmap_diff_data = fmap_diff_img.get_data()
        
    for i in range(sp[0]):
        for j in range(sp[1]):
            for k in range(sp[2]):
                new_data.append([base_runs[m], bp[m][0]+"-"+bp[m][1], i, j, k, fmap_data[i][j][k], fmap_diff_data[i][j][k], theory_data[i][j][k]])
```

```python
df_theory = pd.DataFrame(new_data, columns = ['m', 'diff', 'i', 'j', 'k', 'fmap_mean', 'fmap_diff', 'theory'])
df_theory.info()
```

```python
idx = df_theory[(df_theory['fmap_mean']==0)&(df_theory['fmap_diff']==0.0)&(df_theory['theory']==0.0)].index
df_theory.drop(idx, inplace=True)
df_theory.reset_index(drop=True, inplace=True)
```
