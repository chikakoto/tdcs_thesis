```python
def create_mask(sub_path, base_runs):
    for m in range(len(base_runs)):
        cwd_path = sub_path+"NIFTI/bmaps_numpy/"
        for n in range(len(base_runs[m])):
            bdir = cwd_path+"base_run_"+base_runs[m][n]+"/"
            print commands.getoutput("fslmaths "+bdir+"4d_fmap_Mean.nii.gz -div "+bdir+"4d_fmap_Mean.nii.gz "+bdir+"mask.nii.gz")
            print commands.getoutput("3dmaskdump -o "+bdir+"/mask.txt -xyz -nozero "+bdir+"mask.nii.gz")
```

```python
def create_erode_image(sub_path, base_runs, bp, erode = "1"):
    for m in range(len(base_runs)):
        cwd_path = sub_path+"NIFTI/bmaps_numpy/"
        tdir = cwd_path+bp[m][0][0]+"minus"+bp[m][0][1]
        print commands.getoutput("3dmask_tool -input "+tdir+"/mask.nii.gz -prefix "+tdir+"/mask_erode"+erode+".nii.gz -dilate_result -"+erode)
        print commands.getoutput("fslmaths "+tdir+"/mask_erode"+erode+".nii.gz -mul "+tdir+"/4d_fmap_Mean.nii.gz "+tdir+"/diff_fmap_Mean_erode"+erode+".nii.gz")
        print commands.getoutput("fslmaths "+tdir+"/mask_erode"+erode+".nii.gz -mul "+tdir+"/Bvals.nii "+tdir+"/Bvals_erode"+erode+".nii.gz")

        for n in range(len(base_runs[m])):
            bdir = cwd_path+"base_run_"+base_runs[m][n]
            print commands.getoutput("3dmask_tool -input "+bdir+"/mask.nii.gz -prefix "+bdir+"/mask_erode"+erode+".nii.gz -dilate_result -"+erode)
            print commands.getoutput("fslmaths "+bdir+"/mask_erode"+erode+".nii.gz -mul "+bdir+"/4d_fmap_Mean.nii.gz "+bdir+"/diff_fmap_Mean_erode"+erode+".nii.gz")
```

```python
def extract_mean(sub_path, experiment):
    print(sub_path)
    cwd_path = sub_path+"/NIFTI/bmaps_numpy/"
    result = pd.DataFrame()
    for m in range(len(base_runs)):   
        tdir = cwd_path+bp[m][0][0]+"minus"+bp[m][0][1]+"/"
        theory_img = nib.load(tdir+"Bvals_erode1.nii.gz")
        theory_data = theory_img.get_data()
        sp = theory_data.shape
        b_data = []

        for i in range(sp[0]):
            for j in range(sp[1]):
                for k in range(sp[2]):
                    b_data.append([experiment, m+1, i, j, k, theory_data[i][j][k]])

        b_df = pd.DataFrame(b_data, columns=['exp', 'mini_exp', 'i', 'j', 'k', 'theory'])

        df = pd.DataFrame()
        for n in range(len(base_runs[m])):
            bdir = cwd_path+"base_run_"+base_runs[m][n]+"/"
            print(base_runs[m][n])

            new_data = []      
            fmap_img = nib.load(bdir+"diff_fmap_Mean_erode1.nii.gz")
            fmap_data = fmap_img.get_data()
            sp = fmap_data.shape            
            col = "mean"+str(n)

            for i in range(sp[0]):
                for j in range(sp[1]):
                    for k in range(sp[2]):
                        new_data.append([experiment, m+1, i, j, k, fmap_data[i][j][k]])

            new_df = pd.DataFrame(new_data, columns=['exp', 'mini_exp', 'i', 'j', 'k', col])
            if n == 0:
                df = new_df
            else:
                df = df.merge(new_df, on=['exp', 'mini_exp', 'i', 'j', 'k'], sort=False)


        df = df.merge(b_df, on=['exp','mini_exp', 'i', 'j', 'k'], sort=False)
        result = result.append(df, ignore_index=True)
    return result
```

## 2. 3 attributes (separete negative)

```python
def extract_mean_separate_neg(sub_path, experiment):
    print(sub_path)
    cwd_path = sub_path+"/NIFTI/bmaps_numpy/"
    result = pd.DataFrame()
    for m in range(len(base_runs)): # mini experiment
        # Bvals
        for l in range(2):
            tdir = cwd_path+bp[m][l][0]+"minus"+bp[m][l][1]+"/"
            theory_img = nib.load(tdir+"Bvals_erode1.nii.gz")
            theory_data = theory_img.get_data()
            sp = theory_data.shape
            b_data = []
            
            neg = 0
            if l==1:
                neg = 1

            for i in range(sp[0]):
                for j in range(sp[1]):
                    for k in range(sp[2]):
                        b_data.append([experiment, m+1, i, j, k, 2, neg, theory_data[i][j][k]])
            
            if l==0:
                b_df = pd.DataFrame(b_data, columns=['exp', 'mini_exp', 'i', 'j', 'k', 'amp', 'neg', 'theory'])
            else:
                b_df2 = pd.DataFrame(b_data, columns=['exp', 'mini_exp', 'i', 'j', 'k', 'amp', 'neg', 'theory'])

        df = pd.DataFrame()
        df2 = pd.DataFrame()
        new_df = pd.DataFrame()
        new_df2 = pd.DataFrame()
        for n in range(len(base_runs[m])): # 5 runs
            bdir = cwd_path+"base_run_"+base_runs[m][n]+"/"
            print(base_runs[m][n])

            new_data = []    
            new_data2 = [] 
            fmap_img = nib.load(bdir+"diff_fmap_Mean_erode1.nii.gz")
            fmap_data = fmap_img.get_data()
            sp = fmap_data.shape  
       
            for i in range(sp[0]):
                for j in range(sp[1]):
                    for k in range(sp[2]):
                        new_data.append([experiment, m+1, i, j, k, 2, 0, fmap_data[i][j][k]])
                        new_data2.append([experiment, m+1, i, j, k, 2, 1, fmap_data[i][j][k]])

            if n<=2:          
                col = "mean"+str(n)
                new_df = pd.DataFrame(new_data, columns=['exp', 'mini_exp', 'i', 'j', 'k', 'amp', 'neg', col])
            if n>=2:
                col = "mean"+str(n-2)
                new_df2 = pd.DataFrame(new_data2, columns=['exp', 'mini_exp', 'i', 'j', 'k', 'amp', 'neg', col])
                
            if n == 0:
                df = new_df 
            elif n<=2:
                df = df.merge(new_df, on=['exp', 'mini_exp', 'i', 'j', 'k', 'amp', 'neg'], sort=False)

            if n == 2:
                df2 = new_df2
            elif n>2:
                df2 = df2.merge(new_df2, on=['exp', 'mini_exp', 'i', 'j', 'k', 'amp', 'neg'], sort=False)

        df = df.merge(b_df, on=['exp','mini_exp', 'i', 'j', 'k', 'amp', 'neg'], sort=False)
        df2 = df2.merge(b_df2, on=['exp','mini_exp', 'i', 'j', 'k', 'amp', 'neg'], sort=False)
        
        result = result.append(df, ignore_index=True)
        result = result.append(df2, ignore_index=True)
    return result
```
